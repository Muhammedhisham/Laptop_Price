```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pickle
```

```python
data=pd.read_csv('laptopData.csv')
```

```python
data.head()
```

```python
data.drop(['Unnamed: 0'],axis=1,inplace=True)
data.drop(['Weight'],axis=1,inplace=True)
```

```python
data.head(2)
```

```python
data.dropna(inplace=True)
```

```python
data.info()
```

```python
data['Memory'].unique()
```

```python
first=[]
second=[]
for s in data['Memory']:
    if '+' in s:
        mdata=s.split(' +  ')
        print(mdata)
        fdata=mdata[0]
        ldata=mdata[1]
    else:
        fdata=s
        ldata='No additional storage'
    first.append(fdata)
    second.append(ldata)
```

```python
data['Memory1']=first
data['Memory2']=second
data.drop(['Memory'],axis=1,inplace=True)
```

```python
data.head()
```

```python
Processor=[]
for cp in data['Cpu']:
    if 'i3' in cp:
        cp='Intel Core i3'
    elif 'i5' in cp:
        cp='Intel Core i5'
    elif 'i7' in cp:
        cp='Intel Core i7'
    elif 'Celeron' in cp:
        cp='Intel Celeron'
    elif 'Pentium' in cp:
        cp='Intel Pentium'
    elif 'Xeon' in cp:
        cp='Intel Xeon'
    elif 'Core M' in cp:
        cp='Intel Core M'
    elif 'Atom' in cp:
        cp='Intel Atom'
    elif 'Samsung Cortex' in cp:
        cp='Samsung Cortex'
    elif 'AMD E-Series' in cp:
        cp='AMD E-Series'
    elif 'A9-Series' in cp:
        cp='AMD A9-Series'
    elif 'AMD A10-Series' in cp:
        cp='AMD A10-Series'
    elif 'AMD A12-Series' in cp:
        cp='AMD A12-Series'
    elif 'A4-Series' in cp:
        cp='AMD A4-Series'
    elif 'AMD A6-Series' in cp:
        cp='AMD A6-Series'
    elif 'AMD A8-Series' in cp:
        cp='AMD A8-Series'
    elif 'AMD FX' in cp:
        cp='AMD FX'
    elif 'AMD Ryzen' in cp:
        cp='AMD Ryzen'
    else:
        cp=cp
    Processor.append(cp)   
Processor     
```

```python
data['Processor']=Processor
```

```python
data.drop(['Cpu'],axis=1,inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder=LabelEncoder()
```

```python
data['Company']=labelencoder.fit_transform(data['Company'])
data['TypeName']=labelencoder.fit_transform(data['TypeName'])
data['ScreenResolution']=labelencoder.fit_transform(data['ScreenResolution'])
data['Processor']=labelencoder.fit_transform(data['Processor'])
data['Memory1']=labelencoder.fit_transform(data['Memory1'])
data['Memory2']=labelencoder.fit_transform(data['Memory2'])
data['Gpu']=labelencoder.fit_transform(data['Gpu'])
data['OpSys']=labelencoder.fit_transform(data['OpSys'])
```

```python
data
```

```python
y=data['Price']
x=data.drop(['Price'],axis=1)
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest =train_test_split(x,y,test_size=0.3)
```

```python
from sklearn.ensemble import RandomForestRegressor
RF=RandomForestRegressor()
model_RF=RF.fit(xtrain,ytrain)
model_RF.score(xtest,ytest)
```

```python
model_RF.predict(xtest)
```

```python
from sklearn.tree import DecisionTreeRegressor
DT = DecisionTreeRegressor()
model_DT = DT.fit(xtrain, ytrain)
model_DT.score(xtest, ytest)

```

```python
with open('output.pkl', 'wb') as f:
    pickle.dump(model_RF, f)
```
